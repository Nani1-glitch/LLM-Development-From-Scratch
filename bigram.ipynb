```python
import torch # Helps in doing lot of calculus and leniar algebra and also has tensors
device = 'cuda' if torch.cuda.is_available() else 'cpu' # Checks whether GPU is available or not 
print(device)

# Called as Hyper parameters 
block_size = 8
batch_size = 4 
```

```python
with open('wizard_of_oz.txt', 'r', encoding = 'utf-8') as f:
    text = f.read()
chars = sorted(set(text))
print(chars)
print(len(chars))
```

```python
string_to_int = {ch:i for i,ch in enumerate(chars)}
int_to_string = {i:ch for i,ch in enumerate(chars)}
encode = lambda s: [string_to_int[c] for c in s]
decode = lambda l: ''.join([int_to_string[i] for i in l])

# Before importing the torch
encoded_hello = encode('hello') # This converts the text/string into the integers
decoded_hello = decode(encoded_hello) # This decodes the output generated by the above line.      
# The process of encoding the string to text and text to string has been done in the above two lines  

print(decoded_hello)
```

```python
# Write now we are working on small amount of vocabulary and large amount of tokenizers 
```

```python
string_to_int = {ch:i for i,ch in enumerate(chars)}
int_to_string = {i:ch for i,ch in enumerate(chars)}
encode = lambda s: [string_to_int[c] for c in s]
decode = lambda l: ''.join([int_to_string[i] for i in l])

data = torch.tensor(encode(text), dtype = torch.long) # Tensors can be used to dp math things and perform the calculations 
print(data[:100])
```

```python
n = int(0.8 * len(data))
train_data = data[:n]
val_data = data[n:]
```

```python
block_size = 8

x = train_data[:block_size] # Here x is our production 
y = train_data[1:block_size + 1] # y is our targets 

for t in range(block_size): # To show what the predictions and targets are 
    context = x[:t+1]
    target = y[t]
    print('when input is', context, 'target is', target)
```

# We can stack these blocks like one on one which resemble like a stack data structure. This generates the hyper parameter known as BATCH.
  Block size is length of each sequence like [ . . . . . . . . . . . . . ]
  Batch size is how many of these blocks that we process at the same time.

  This is the very good way to train the language models.
